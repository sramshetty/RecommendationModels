```python
f = open('./movies.txt', 'r', encoding='ISO-8859-1')
cycle_line_num = 8
line_index = 0
# beer_df = pd.DataFrame(columns=['userid', 'itemid', 'timestamp', 'category'])
# category_list = []
# beer_id_list = []
user_id = -1
beer_id = -1
timestamp = -1
category = -1
global_line_index = 0
data_dict = {}

data_dict['userid'] = []
data_dict['itemid'] = []
data_dict['timestamp'] = []
# data_dict['category'] = []

line_num = 0

for line in f:
    
    if "product/" in line:
        line_index += 1
    elif "review/" in line:
        line_index += 1
    else:
        line_index += 0
        
#     print(line_index, "line", line)
        
    line_num += 1
    
#     if line_num > 1274200:
#         print("line", line)
    
    if line_index == cycle_line_num:
        line_index = 0
        
        if int(timestamp) < 1262304000:
            continue
#         print("timestamp", timestamp)
        global_line_index += 1
        data_dict['userid'].append(user_id)
        data_dict['itemid'].append(beer_id)
        data_dict['timestamp'].append(timestamp)
        if global_line_index %5e6 == 0 and line_index == 1:
            print("global_line_index", global_line_index)
#         data_dict['category'].append(category)
#         beer_df.loc[global_line_index] = [user_id, beer_id, timestamp, category
#         data_dict[global_line_index] = {"userid": user_id, "itemid":beer_id, "timestamp": timestamp, "category": category}
    
    if line_index == 1:
        beer_id = line.split(":")[1]
        
#     if line_index == 5:
#         category = line.split(":")[1]
        
    if line_index == 6:
        timestamp = line.split(":")[1]  
       
    if line_index == 2:
        user_id = line.split(":")[1]
        
    if line_num % 1e7 == 0:
        print("line_num", line_num)
        
#     if global_line_index > 1e6:
#         break
f.close()
```

```python
len(data_dict['userid']), len(data_dict['itemid']), len(data_dict['timestamp'])
```

```python
df = pd.DataFrame.from_dict(data_dict)
df.columns=["userid", "itemid", "timestamp"]
```

```python
movie_df = df
movie_df['userid'] = df['userid'].map(lambda x: x.lstrip(" ").rstrip('\n'))
movie_df['itemid'] = df['itemid'].map(lambda x: x.lstrip(" ").rstrip('\n'))
movie_df['timestamp'] = df['timestamp'].map(lambda x: x.lstrip(" ").rstrip('\n'))
```

```python
time_sorted_df = movie_df.sort_values(by=['userid', 'timestamp'], ascending=[True, True])
```
